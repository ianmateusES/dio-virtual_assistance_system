```python
import os
from datetime import datetime
```

## Text-to-Speech

```python
import librosa
import soundfile as sf
from gtts import gTTS
from IPython.display import Audio
```

```python
# Criando o diretorio que vai conter os audios
path = '/content/Speech'
if not os.path.exists(path):
    os.mkdir(path)
```

```python
text = 'Que dia Ã© hoje?'
lang = 'pt'

speech = gTTS(text=text, lang=lang, slow=True)
name_speech = f'{datetime.now().timestamp()}.wav'

speech.save(f'./Speech/{name_speech}')

x,_ = librosa.load(f'/content/Speech/{name_speech}', sr=16000)
sf.write(f'/content/Speech/{name_speech}', x, 16000)

del x, speech
```

```python
Audio(f'/content/Speech/{name_speech}')
```

```python
import webbrowser
import speech_recognition as sr
from pydub import AudioSegment
```

```python
files = [i for i in os.listdir('/content/Speech') if i != '.ipynb_checkpoints']
files.sort(reverse=True)
files
```

```python
lang = 'pt'
name_speech = files[0]

r = sr.Recognizer()

with sr.AudioFile(f'/content/Speech/{name_speech}') as source:
    audio = r.record(source)  # LÃª o Ã¡udio do arquivo

try:
    transcript = r.recognize_google(audio, language=lang)
    print("TranscriÃ§Ã£o: " + transcript)
    url = f"https://www.google.com/search?q={transcript}"
    webbrowser.open(url)
except sr.UnknownValueError:
    print("NÃ£o foi possÃ­vel reconhecer o Ã¡udio.")
except sr.RequestError:
    print("Erro na solicitaÃ§Ã£o ao serviÃ§o de reconhecimento de fala.")
```
